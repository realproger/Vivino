```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_wine(csv_d):
    data_p = pd.read_csv(csv_d, sep=',')
    
    return data_p
data_w = load_wine("wines.csv")
```

```python
def clean_wine(ddt):
    d_clean = ddt.dropna()

    d_clean["type"] = d_clean["type"].map({"white": 0, "red": 1})

#     totalization = ddt.isna().sum()
    
    return d_clean
cleaned_wine = clean_wine(data_w)
cleaned_wine
```

```python
def analyze_wine_data(data):
    
    # Visualization of histograms
    data.hist(bins=50, figsize=(8, 12))
    plt.show()
    
    # Calculate correlation and build a heatmap
    cor_mtrx = data.drop(columns=['type']).corr()
    plt.figure(figsize=(10, 8))
    sns.set(font_scale=1.2)
    sns.heatmap(cor_mtrx, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
    plt.title("Correlation Heatmap")
    plt.show()
    
    # Plotting the pairplo
    sns.pairplot(data.drop(columns=['type']))
    plt.show()
    
analyze_wine_data(cleaned_wine)
```

```python
def wine_recommendation(a_data):
    features = a_data.iloc[:, 1:]
    target = a_data.iloc[:, 0]
    
#     threshold = 4.5
#     target_binary = target.apply(lambda x: 1 if  x >= threshold else 0)
    
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
    
    # For data standardization
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Object of the Logistic Regression classifier
    model = LogisticRegression()
    model.fit(X_train_scaled, y_train)
    predictions = model.predict(X_test_scaled)
    
    # Percentage accuracy of predictions
    accuracy_sc = accuracy_score(y_test, predictions) * 100
    print(f"Accuracy - {accuracy_sc:.2f}%")
    
    cmp_values = pd.DataFrame({'True Label': y_test.squeeze(), 'Predicted Label': predictions.squeeze()})
    pd.set_option('display.max_rows', cmp_values.shape[0]+1)
    return cmp_values.iloc[:251]

wine_recs = wine_recommendation(cleaned_wine)
wine_recs
```

```python
# Serves to evaluate the performance of the classifier on the test data
def evaluate_classifier(X_train, X_test, y_train, y_test, clf):
    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_test)
    
    return r2_score(y_test, y_pred), accuracy_score(y_test, y_pred) # calculated r2 and accuracy score
```

```python
def wine_prediction(prep_data):
    
    # splitting he dataset info features(X) and target labels(y)
    X = prep_data.iloc[:, :-1]
    y = prep_data.iloc[:, -1]
    
    #training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
    
    # classifier instances
    rf_clf = RandomForestClassifier()
    lda_clf = LinearDiscriminantAnalysis()
    knn_clf = KNeighborsClassifier()
    
    # evaluating and printing results for ...
    print("random forest classifier: ")
    randf_r2, randf_acc = evaluate_classifier(X_train, X_test, y_train, y_test, rf_clf)
    print(f"R2 score: {randf_r2 * 100: .2f}")
    print(f"Accuracy score: {randf_acc * 100: .2f}")
    
    
    print("\nlinear discriminant analysis: ")
    lda_r2, lda_acc = evaluate_classifier(X_train, X_test, y_train, y_test, lda_clf)
    print(f"R2 score: {lda_r2 * 100: .2f}")
    print(f"Accuracy score: {lda_acc * 100: .2f}")
    
    
    print("\nk-nearest neighbors: ")
    knn_r2, knn_acc = evaluate_classifier(X_train, X_test, y_train , y_test, knn_clf)
    print(f"R2 score: {knn_r2 * 100: .2f}")
    print(f"Accuracy score: {knn_acc * 100: .2f}")
    
    
          
wine_prediction(cleaned_wine)
```
